**Q.2 How does broadcasting work in NumPy?**

**Broadcasting in NumPy**
Broadcasting is a feature in NumPy that allows operations between arrays of different shapes without explicitly reshaping or replicating data. It automatically expands smaller arrays to match the shape of larger arrays, making computations more efficient.

**How Broadcasting Works**
When performing arithmetic operations on arrays, NumPy follows these rules:

1. **If the arrays have the same shape**, element-wise operations occur normally.
2. **If the arrays have different shapes**, NumPy tries to "broadcast" the smaller array to match the shape of the larger one by expanding dimensions.
3. **Broadcasting Rules**:
   - The dimensions are compared **from right to left**.
   - If the dimensions are equal, they are compatible.
   - If one dimension is '1', it is stretched to match the other dimension.
   - If the dimensions are incompatible, NumPy raises a **ValueError**.

**Examples**
**Example 1: Adding a scalar to an array**
```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
scalar = 10

result = arr + scalar  # Scalar is broadcasted across all elements
print(result)
```
**Output:**
```
[[11 12 13]
 [14 15 16]]
```

**Example 2: Broadcasting a 1D array to a 2D array**
```python
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr_1d = np.array([10, 20, 30])

result = arr_2d + arr_1d  # arr_1d is broadcasted across rows
print(result)
```
**Output:**
```
[[11 22 33]
 [14 25 36]]
```

**Q4. Explain the use of the groupby() method in Pandas.**

**`groupby()` Method in Pandas**
The **`groupby()`** method in Pandas is used to group data based on one or more columns and perform aggregate functions like **sum**, **mean**, **count**, **max**, **min**, etc., on the grouped data.

It works similarly to the **GROUP BY** clause in SQL.

---

**Syntax**

DataFrame.groupby(by, axis=0).agg(function)

- **by**: Column(s) to group by (string or list of column names)
- **axis**: Default is `0` (group by rows)
- **agg(function)**: Aggregate function to apply on the grouped data (like `sum()`, `mean()`, etc.)

---

**Example 1: Grouping by a Single Column**
Let's group employees by their department and calculate their average salary.
```python
import pandas as pd

data = {
    'Department': ['HR', 'IT', 'HR', 'Finance', 'IT'],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [40000, 60000, 45000, 70000, 65000]
}

df = pd.DataFrame(data)

# Group by Department and calculate average salary
grouped = df.groupby('Department')['Salary'].mean()
print(grouped)
```
**Output:**
```
Department
Finance    70000.0
HR         42500.0
IT         62500.0
Name: Salary, dtype: float64
```

**Q6.What are the differences between NumPy arrays and Python lists?**
  

| Feature            | **NumPy Arrays** | **Python Lists** |
|--------------------|-------------------|--------------------|
| **Performance**   | Faster (C-optimized, contiguous memory) | Slower (dynamic, heterogeneous types) |
| **Memory Usage**  | Uses less memory (fixed data types) | Uses more memory (stores references) |
| **Functionality** | Supports mathematical and vectorized operations | Requires loops for element-wise operations |
| **Data Type**     | Homogeneous (all elements must be of the same type) | Heterogeneous (can contain mixed data types) |
| **Multi-dimensional Support** | Supports n-dimensional arrays (e.g., 2D, 3D) | Only supports 1D lists (nested lists for higher dimensions) |
| **Built-in Operations** | Optimized mathematical functions | No built-in support for element-wise math |
| **Ease of Use**   | More suited for numerical computing and large datasets | More general-purpose and flexible |

---

**1. Performance Comparison**

NumPy arrays are **faster** because they store elements in **contiguous memory blocks** and use optimized C-based functions.  

**Example: NumPy vs. List Performance**

```python
import numpy as np
import time

size = 10**6  # 1 million elements

# Python List
py_list = list(range(size))
start = time.time()
py_list = [x * 2 for x in py_list]  # Loop-based operation
print("Python List Time:", time.time() - start)

# NumPy Array
np_array = np.arange(size)
start = time.time()
np_array = np_array * 2  # Vectorized operation
print("NumPy Array Time:", time.time() - start)
```
👉 **NumPy is significantly faster! 🚀**

---

**2. Memory Usage**

NumPy arrays use **less memory** because they store data in a **fixed type** rather than storing references.  

**Example: Checking Memory Size**
```python
import sys

py_list = [1, 2, 3, 4, 5]
np_array = np.array([1, 2, 3, 4, 5])

print("Python List Memory:", sys.getsizeof(py_list))  # More memory due to object references
print("NumPy Array Memory:", np_array.nbytes)  # Less memory due to fixed type
```

---

**3. Multi-Dimensional Support**

NumPy supports **multi-dimensional** arrays, while Python lists only support **1D** lists (nested lists are used for 2D or 3D).  

**Example: 2D NumPy Array vs. Nested List**
```python
# 2D NumPy Array
np_arr = np.array([[1, 2, 3], [4, 5, 6]])
print(np_arr.shape)  # (2,3)

# Nested Python List
py_list = [[1, 2, 3], [4, 5, 6]]
print(len(py_list), len(py_list[0]))  # Requires manual shape handling
```

---

**4. Mathematical Operations**
NumPy **supports vectorized operations**, while Python lists require **loops**.

**Example: Squaring Elements**
```python
# Python List (Loop Required)
py_list = [1, 2, 3, 4]
squared_list = [x**2 for x in py_list]

# NumPy Array (Vectorized)
np_array = np.array([1, 2, 3, 4])
squared_array = np_array**2  # Faster!

print(squared_list)
print(squared_array)
```

---

**5. Built-in Functions**

NumPy has **many built-in mathematical functions**, whereas Python lists require **manual implementation**.

**Example: Sum of Elements**
```python
# Python List
py_list = [1, 2, 3, 4]
print(sum(py_list))  # Slower

# NumPy Array
np_array = np.array([1, 2, 3, 4])
print(np_array.sum())  # Faster!
```



**Q8.What is a "Vectorized Operation" in NumPy?**


A **vectorized operation** in NumPy refers to performing operations on entire arrays (vectors) **without using explicit loops**. These operations are highly optimized and execute much faster than traditional Python loops.


**Why Use Vectorized Operations?**

**Faster Execution:** Uses optimized C-based implementations.  
**Less Code:** No need for `for` loops.  
**More Readable:** Operations look like simple mathematical expressions.  
**Memory Efficient:** Reduces unnecessary temporary variables.  

---

**Examples of Vectorized Operations**

**1. Basic Arithmetic**
```python
arr = np.array([1, 2, 3, 4, 5])

print(arr + 2)  # Add 2 to each element
print(arr * 3)  # Multiply each element by 3
print(arr ** 2) # Square each element
```

**2. Applying Mathematical Functions**

```python
print(np.sin(arr))   # Compute sine of each element
print(np.sqrt(arr))  # Compute square root
print(np.log(arr))   # Compute natural logarithm
```

**3. Vectorized Dot Product (Faster Matrix Multiplication)**

```python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

# Without Vectorization (Loop-based)
dot_product = sum(A[i] * B[i] for i in range(len(A)))

# With Vectorization (NumPy)
dot_product_np = np.dot(A, B)

print(dot_product, dot_product_np)  # Same result, but NumPy is faster!
```

**Q.17 Explain the difference between apply() and map() in Pandas.**


Both `apply()` and `map()` are used for applying functions to Pandas objects, but they have different use cases and scopes.  

| Feature | `apply()` | `map()` |
|---------|----------|---------|
| **Applies to** | DataFrame & Series | Only Series |
| **Works on** | Rows or columns (DataFrame) & elements (Series) | Element-wise operations on Series |
| **Function Type** | Can apply custom or built-in functions | Works with functions, dictionaries, or Series |
| **Returns** | Series, DataFrame, or scalar | Series |

---

**1. `apply()` – Works on DataFrame & Series**  
Used for applying **functions to Series and DataFrame rows/columns**.  

**Example: Using `apply()` on a Series**
```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3, 4]})

# Apply a function to double each value
df['A_doubled'] = df['A'].apply(lambda x: x * 2)
print(df)
```
**Output:**
```
   A  A_doubled
0  1         2
1  2         4
2  3         6
3  4         8
```

**Example: Using `apply()` on a DataFrame**
```python
# Apply a function on all columns
df['Sum'] = df.apply(lambda row: row.sum(), axis=1)
```
---

**2. `map()` – Works Only on Series (Element-wise Transformation)**  
Used for applying functions or mapping values **only to Series**.

**Example: Using `map()` for Element-wise Transformation**
```python
df['A_squared'] = df['A'].map(lambda x: x ** 2)
print(df)
```
---

**Key Differences in Practice**

| Use Case | Preferred Method |
|----------|-----------------|
| Applying a function to a **DataFrame** | `apply()` |
| Applying a function to a **Series** | Both `apply()` and `map()` |
| Using a dictionary or mapping values in a Series | `map()` |

For DataFrames, always use `apply()`, while for Series, `map()` is more efficient for element-wise transformations.

**Q.21 Why is NumPy’s array slicing faster than Python’s list slicing?**


NumPy’s array slicing is significantly faster than Python’s list slicing because of **efficient memory management and optimized operations**. Here’s why:  

---

**1. NumPy Uses Contiguous Memory Allocation**  
- NumPy arrays are stored in **a single contiguous block of memory**, unlike Python lists, which store references to objects.  
- This allows NumPy to access elements quickly without following multiple memory pointers.  

🔹 **Example: Checking memory locations**  
```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr.data)  # Shows contiguous memory location
```
---

**2. Slicing in NumPy Creates Views, Not Copies**  
- In NumPy, slicing returns a **view** of the original array, not a new copy.  
- This saves memory and avoids extra computations.  

🔹 **Example: NumPy Slicing (View, No Copy)**  
```python
arr = np.array([1, 2, 3, 4, 5])
sliced_arr = arr[1:4]  # No new array is created
sliced_arr[0] = 99  # Modifies original array
print(arr)  # Output: [ 1 99  3  4  5 ]
```
---

**3. Python Lists Store References (More Overhead)**  
- Python lists store elements as **pointers to objects**, so slicing involves **creating a new list** with new references.  
- This is slower because it requires additional memory allocation and copying.  

🔹 **Example: Python List Slicing (Creates New Copy)**  
```python
lst = [1, 2, 3, 4, 5]
sliced_lst = lst[1:4]  # Creates a new list
sliced_lst[0] = 99
print(lst)  # Output: [1, 2, 3, 4, 5] (Original list unchanged)
```
---

**4. NumPy Uses Optimized C/Fortran Implementations**  
- NumPy is implemented in **C and Fortran**, making operations like slicing much faster than Python’s built-in list operations.  
- Python lists require **interpreted loops**, which are slower than NumPy’s vectorized operations.  

🔹 **Example: Performance Comparison**  





````python
import time

lst = list(range(1000000))
arr = np.arange(1000000)

# Python list slicing
start = time.time()
lst_slice = lst[:500000]
end = time.time()
print("List slicing time:", end - start)

# NumPy array slicing
start = time.time()
arr_slice = arr[:500000]
end = time.time()
print("NumPy slicing time:", end - start)
````
